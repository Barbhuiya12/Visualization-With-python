Ex.3 Stack Bar Graph 

```python
labels = ['Test 1', 'Test 2', 'Test 3', 'Test 4', 'Test 5', 'Test 6',]

group_one_price = [30, 40, 50, 50, 30, 20]
group_two_price = [10, 20, 30, 40, 50, 20]
group_one_tax = [1,2,1,2,1,1]
group_two_tax = [3,4,0,4,3,4]

width = 0.50
fig, ax = plt.subplots()

#create our stacked bar visualization
ax.bar(labels, group_one_price, width, align = 'edge', yerr=group_one_tax, label='First Label')
ax.bar(labels, group_two_price, width, align = 'edge', yerr=group_two_tax, bottom=group_one_price, label='Second Label')

ax.set_ylabel('Value')
ax.set_title('Scores by Group')
ax.legend()
plt.show()
```

Different style By using ggplot

```python
plt.style.use('ggplot')
labels = ['Test 1', 'Test 2', 'Test 3', 'Test 4', 'Test 5', 'Test 6',]

group_one_price = [30, 40, 50, 50, 30, 20]
group_two_price = [10, 20, 30, 40, 50, 20]
group_one_tax = [1,2,1,2,1,1]
group_two_tax = [3,4,3,4,3,4]

width = 0.50
fig, ax = plt.subplots()

#create our stacked bar visualization
ax.bar(labels, group_one_price, width, align = 'edge', yerr=group_one_tax, label='First Label')
ax.bar(labels, group_two_price, width, align = 'edge', yerr=group_two_tax, bottom=group_one_price, label='Second Label')###Explore yerr

ax.set_ylabel('Value')
ax.set_title('Scores by Group')
plt.xticks(rotation=90)
ax.legend()
plt.show()
```

Create the subplots 

```python
plt.style.use("ggplot")
data = {'Chocolate':20, "Vanilla":10, 'Coffee': 25, 'Pistachio': 15}
names = list(data.keys())
values = list(data.values())
fig, axs = plt.subplots(1,3, figsize=(5,5), sharey=True)###Share(The axis is sahre among the plots)
axs[0].bar(names,values)
axs[1].scatter(names,values)
axs[2].plot(names,values)
fig.suptitle('Plotting')
plt.xticks(rotation=90)
```

Ex 4 Single sub plote 

```python
plt.style.use("ggplot")
x = np.linspace(2,10 * np.pi, 20)
y = np.sin(x ** 7)
fig, ax = plt.subplots()
ax.plot(x,y,color='black')
ax.set_title('A Single Plot')
```

By using tableau color Blind 

```python
plt.style.use('tableau-colorblind10')

x = np.linspace(2,10 * np.pi, 20)
y = np.sin(x ** 7)

fig, axs = plt.subplots(2)
axs[0].plot(x,y, color='black')
axs[1].plot(x,y**2)
```
